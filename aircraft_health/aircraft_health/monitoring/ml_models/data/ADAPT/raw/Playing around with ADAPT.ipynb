```python
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

# Define the data directory as the current directory
data_dir = Path('.')

# List available files in the current directory
print("Available files:")
for file in data_dir.glob('*'):
    print(file.name)

# Path to the raw dataset
file_path = data_dir / 'Exp_304_comp3_pb.txt'

# Load the raw dataset, skipping the first line (metadata)
try:
    df = pd.read_csv(file_path, delim_whitespace=True, skiprows=1)
    
    # Display basic information
    print("\nDataFrame Info:")
    print(df.info())
    
    # Perform descriptive statistics
    print("\nDescriptive Statistics:")
    print(df.describe())
    
    # Example visualization: Plot a specific column
    plt.figure(figsize=(10, 6))
    column = 'E135'  # Replace with an actual column name
    if column in df.columns:
        df[column].hist()
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.title(f'Histogram of {column}')
        plt.show()
    else:
        print(f"Column '{column}' not found in DataFrame.")
except FileNotFoundError:
    print(f"File '{file_path}' not found. Please check the file path and ensure the file exists.")
```
